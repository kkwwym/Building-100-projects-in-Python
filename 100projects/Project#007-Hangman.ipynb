```python
import random
```

```python
stages = ['''
  +---+
  |   |
  O   |
 /|\  |
 / \  |
      |
=========
''', '''
  +---+
  |   |
  O   |
 /|\  |
 /    |
      |
=========
''', '''
  +---+
  |   |
  O   |
 /|\  |
      |
      |
=========
''', '''
  +---+
  |   |
  O   |
 /|   |
      |
      |
=========''', '''
  +---+
  |   |
  O   |
  |   |
      |
      |
=========
''', '''
  +---+
  |   |
  O   |
      |
      |
      |
=========
''', '''
  +---+
  |   |
      |
      |
      |
      |
=========
''']
```

```python
len(stages)
```

```python
word_list=["complex","baboon","camel","apple","python","water","solution"]
```

```python
#Randomly choose a word
chosen_word=random.choice(word_list)
word_length=len(chosen_word)

#Set the rule of game at first 
end_of_game=False


#Give user a hint:How many words we have to guess
display=[]
for _ in range(word_length):
    display.append("_")
print(display)

games_count=0
print(stages[games_count])



#Ask the user to guess a letter
while not end_of_game:
    guess=input("Guess a letter: ").lower()
    #games_count+=1
    for position in range(word_length):
        letter=chosen_word[position]
        if letter==guess:
            display[position]=letter
    if guess in display:
        print(stages[games_count])
    else:
        games_count+=1
            
        if games_count<=5:
            print(stages[games_count])
        else:
            end_of_game=True
            print(stages[games_count])
            print("You die")
            
    print(display)
    
    
    
    if "_" not in display:
        end_of_game=True
        print("You Win")
```
